```python
from PIL import Image
import gtts
import pytesseract
```

```python
img = Image.open('sample.png')
img
```

```python
text = pytesseract.image_to_string(img)
text
```

```python
tts = gtts.gTTS(text)
tts.save('sample.mp3')
```

```python
from IPython.display import Audio
Audio("sample.mp3", autoplay=True)
```
